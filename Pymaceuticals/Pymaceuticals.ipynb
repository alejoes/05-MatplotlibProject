```python
placeb=["a262","a897","c282","c757","c766","e227","o795","q787","t718","x773","y478","i477","i669","j235","l872",
            "n763","o302","q582","r850","s152","s166","t994","v409","v989","w167"]
capom=["b128","b742","g288","g316","i557","i738","j119","l509","l897","m601","m957","r554",
        "r944","s185","s710","t565","u364","v923","w914","x401","y793","f966","j246","r157","w150"]
ketap=["c832","g497","h246","k382","m550","m650","o331","p189","u327","w422","y260","v603","v289","q119","o973",
        "n923","l264","h428","g107","f932","f278","d474","c819","c580","a457"]
infu=["a203","a251","a685","c139","e476","f345","k483","y163","z581","a577","c326","c895","i386","k804","m756",
        "n671","o809","o813","q132","s121","v339","v719","v766","w193","w584"]

def data_d(superT,drug,grad_color,Variab,arrayx,a,b):
    dm = sns.light_palette(grad_color,as_cmap=True,reverse=False)
    data_drug=superT[superT.Drug==drug]
     #Tumor Volume
    data_drug_p=(data_drug.pivot(index="Timepoint",columns="Mouse_ID",values=Variab)).reset_index()
    data_sort=data_drug_p[arrayx]
    Data_Style=data_sort.style.background_gradient(cmap=dm)
         #Average Tumor growth & Standard Deviation    
    AvP=data_drug_p.drop(["Timepoint"],axis=1)  
    Stats=AvP.apply(pd.DataFrame.describe, axis=1)
    Stats["Std_Error"]=Stats["std"]/np.sqrt(AvP.shape[1])
    Stats=Stats.drop(["min","25%","50%","75%","max","std"],axis=1)
    days=data_drug_p["Timepoint"].tolist()
    Stats["Time (days)"]=days
    Stats=Stats.set_index("Time (days)")
    Stats=Stats.rename(columns={"mean":drug+" "+a,"Std_Error":drug+" "+b,"count":drug+" "+"(Alive Mice)"})
    return Stats,data_drug_p,Data_Style

SuperD_P=data_d(superT,"Placebo","red","Tumor Volume (mm3)",placeb,a="(Mean)",b="(StdE)")
SuperD_C=data_d(superT,"Capomulin","green","Tumor Volume (mm3)",capom,a="(Mean)",b="(StdE)")
SuperD_K=data_d(superT,"Ketapril","blue","Tumor Volume (mm3)",ketap,a="(Mean)",b="(StdE)")
SuperD_I=data_d(superT,"Infubinol","yellow","Tumor Volume (mm3)",infu,a="(Mean)",b="(StdE)")

SuperD_mP=data_d(superT,"Placebo","red","Metastatic Sites",placeb,a="(Mean)",b="(StdE)")
SuperD_mC=data_d(superT,"Capomulin","green","Metastatic Sites",capom,a="(Mean)",b="(StdE)")
SuperD_mK=data_d(superT,"Ketapril","blue","Metastatic Sites",ketap,a="(Mean)",b="(StdE)")
SuperD_mI=data_d(superT,"Infubinol","yellow","Metastatic Sites",infu,a="(Mean)",b="(StdE)")
```
